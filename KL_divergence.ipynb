### KL Divergence 
Kullback–Leibler divergence(쿨백-라이블러 발산)

```python
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (4,4) # Make the figures a bit bigger
plt.style.use('fivethirtyeight') 
```

```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import tensorflow as tf
import seaborn as sns
sns.set()
```

$D_{KL}(P\|Q) = \sum_{x\in \chi}P(x)\log_b\left(\frac{P(x)}{Q(x)}\right)$
 
log base인 $b$ 는 2, 10 혹은 $e$ 중 하나로 보통 사용

```python
def kl_divergence(p, q):
    return np.sum(np.where(p != 0, p * np.log(p / q), 0))
```

```python
x = np.arange(-10, 10, 0.001)
p = norm.pdf(x, 0, 2)
q = norm.pdf(x, 1, 2.5)
print(p.shape)
print(q.shape)
```

```python
kl = kl_divergence(p, q)
print(kl)
```

```python
plt.figure(dpi=96)
plt.title('KL(P||Q) = {}'.format(kl))
plt.plot(x, p)
plt.plot(x, q, c='red')
```

```python
kl = kl_divergence(p, q)
print(kl)
```

```python
kl = kl_divergence(q, p)
print(kl)
```
